```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')

# Генерируем случайные данные (в данном случае, броски кубика)
np.random.seed(42)
data = np.random.randint(1,7,size=1000)

# Функция для создания Bootstrap-выборок
def bootstrap_sample(data):
    return np.random.choice(data, size=len(data), replace=True)

# Функция для вычисления статистики интереса (в данном случае, среднего значения)
def calculate_statistic(sample):
    return np.mean(sample)

# Число повторений Bootstrap
num_bootstrap_samples = 1000

# Создаем пустой массив для сохранения результатов Bootstrap
bootstrap_means = np.zeros(num_bootstrap_samples)

# Выполняем Bootstrap
for i in range(num_bootstrap_samples):
    bootstrap_sample_data = bootstrap_sample(data)
    bootstrap_means[i] = calculate_statistic(bootstrap_sample_data)

# Распределение средних значений
plt.hist(bootstrap_means, bins=30, edgecolor='black')
plt.title('Bootstrap Distribution of the Mean')
plt.xlabel('Mean Value')
plt.ylabel('Frequency')
plt.show()

# Вычисляем доверительный интервал (95%)
confidence_interval = np.percentile(bootstrap_means, [2.5, 97.5])
print(f"95% Доверительный интервал: {confidence_interval}")
```
